2019_sales_data_m

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and preprocess data
store_sales = pd.read_csv("2019_sales_data.csv")
store_sales['Quantity Ordered'] = pd.to_numeric(store_sales['Quantity Ordered'], errors='coerce')
store_sales['Order Date'] = pd.to_datetime(store_sales['Order Date'], errors='coerce')
store_sales.dropna(subset=['Quantity Ordered', 'Order Date'], inplace=True)
store_sales['Month'] = store_sales['Order Date'].dt.month

# Aggregate sales data by product and month
product_monthly_sales = store_sales.groupby(['Product', 'Month'])['Quantity Ordered'].sum().reset_index()

# Calculate growth rate
product_monthly_sales['Growth Rate'] = product_monthly_sales.groupby('Product')['Quantity Ordered'].pct_change() * 100
product_monthly_sales['Growth Rate'] = product_monthly_sales['Growth Rate'].fillna(0)

# Define thresholds for categories
def categorize_growth(row):
    if row['Growth Rate'] > 50:
        return 'High Growth'
    elif 5 < row['Growth Rate'] <= 20:
        return 'Moderate Growth'
    elif -5 <= row['Growth Rate'] <= 5:
        return 'Stable Growth'
    else:
        return 'Decline'

product_monthly_sales['Growth Category'] = product_monthly_sales.apply(categorize_growth, axis=1)

# Detailed predictions for January 2024
detailed_predictions = []
for product in product_monthly_sales['Product'].unique():
    product_data = product_monthly_sales[product_monthly_sales['Product'] == product]
    recent_data = product_data.iloc[-3:]
    avg_growth_rate = (recent_data['Growth Rate'] * [0.5, 0.3, 0.2]).sum() / sum([0.5, 0.3, 0.2])

    # Determine the trend
    if avg_growth_rate > 20:
        trend = "expected to sell significantly more"
    elif avg_growth_rate < -5:
        trend = "not expected to sell significantly more"
    else:
        trend = "expected to see stable sales"

    # Add prediction with growth rate
    detailed_predictions.append(
        f"{product} is {trend} in January 2024 with a growth rate of {avg_growth_rate:.2f}%."
    )

# Display detailed predictions
print("Detailed Predictions for January 2024:")
for prediction in detailed_predictions:
    print(prediction)

# Visualization of growth trends
plt.figure(figsize=(12, 6))
sns.lineplot(data=product_monthly_sales, x='Month', y='Growth Rate', hue='Product', marker='o')
plt.title("Growth Trends by Product")
plt.xlabel("Month")
plt.ylabel("Growth Rate (%)")
plt.axhline(y=20, color='green', linestyle='--', label='High Growth Threshold')
plt.axhline(y=-5, color='red', linestyle='--', label='Decline Threshold')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and preprocess data
sales_data = pd.read_csv("2023_sales_data.csv")

# Ensure 'Quantity' and 'Date' are properly formatted
sales_data['Quantity'] = pd.to_numeric(sales_data['Quantity'], errors='coerce')
sales_data['Date'] = pd.to_datetime(sales_data['Date'], format='%d/%m/%Y', errors='coerce')

# Drop rows with missing or invalid data
sales_data.dropna(subset=['Quantity', 'Date'], inplace=True)

# Extract month from the 'Date' column
sales_data['Month'] = sales_data['Date'].dt.month

# Aggregate sales data by product and month
product_monthly_sales = sales_data.groupby(['Product_Name', 'Month'])['Quantity'].sum().reset_index()

# Calculate growth rate
product_monthly_sales['Growth Rate'] = product_monthly_sales.groupby('Product_Name')['Quantity'].pct_change() * 100
product_monthly_sales['Growth Rate'] = product_monthly_sales['Growth Rate'].fillna(0)

# Define thresholds for growth categories
def categorize_growth(row):
    if row['Growth Rate'] > 50:
        return 'High Growth'
    elif 5 < row['Growth Rate'] <= 20:
        return 'Moderate Growth'
    elif -5 <= row['Growth Rate'] <= 5:
        return 'Stable Growth'
    else:
        return 'Decline'

product_monthly_sales['Growth Category'] = product_monthly_sales.apply(categorize_growth, axis=1)

# Detailed predictions for January 2024
detailed_predictions = []
for product in product_monthly_sales['Product_Name'].unique():
    product_data = product_monthly_sales[product_monthly_sales['Product_Name'] == product]
    recent_data = product_data.iloc[-3:]  # Use the last 3 months of data
    avg_growth_rate = (recent_data['Growth Rate'] * [0.5, 0.3, 0.2]).sum() / sum([0.5, 0.3, 0.2])

    # Determine the trend
    if avg_growth_rate > 20:
        trend = "expected to sell significantly more"
    elif avg_growth_rate < -5:
        trend = "not expected to sell more"
    else:
        trend = "expected to see stable sales"

    # Add prediction with growth rate
    detailed_predictions.append(
        f"{product} is {trend} in January 2024 with a growth rate of {avg_growth_rate:.2f}%."
    )

# Display detailed predictions
print("Detailed Predictions for January 2024:")
for prediction in detailed_predictions:
    print(prediction)

# Visualization of growth trends
plt.figure(figsize=(12, 6))
sns.lineplot(data=product_monthly_sales, x='Month', y='Growth Rate', hue='Product_Name', marker='o')
plt.title("Growth Trends by Product")
plt.xlabel("Month")
plt.ylabel("Growth Rate (%)")
plt.axhline(y=20, color='green', linestyle='--', label='High Growth Threshold')
plt.axhline(y=-5, color='red', linestyle='--', label='Decline Threshold')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from pandas.tseries.offsets import DateOffset
from statsmodels.tsa.arima.model import ARIMA

# Load the dataset
df = pd.read_csv('2023_sales_data.csv')

# Rename columns (adjust to your column names)
df.columns = ["Invoice_ID", "Customer_ID", "Gender", "Product_Name", "Category",
              "Unit_price", "Quantity", "Tax_5%", "Total", "Date", "Payment_Method"]

# Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')

# Set Date as the index
df.set_index('Date', inplace=True)

# Group by month and sum the Total column
monthly_sales = df['Total'].resample('M').sum()

# Plot the raw data
monthly_sales.plot(figsize=(10, 5), title='Monthly Sales')
plt.show()

# Perform ADFuller test for stationarity
def adfuller_test(series):
    result = adfuller(series)
    print(f"ADF Test Statistic: {result[0]}")
    print(f"p-value: {result[1]}")
    if result[1] <= 0.05:
        print("Data is stationary.")
    else:
        print("Data is non-stationary.")

adfuller_test(monthly_sales)

# Differencing to make data stationary if needed
monthly_sales_diff = monthly_sales.diff().dropna()
adfuller_test(monthly_sales_diff)

# # Fit the ARIMA model
# model = ARIMA(monthly_sales, order=(1, 1, 1))
# results = model.fit()

# # Summary of the model
# print(results.summary())

# # Forecast future sales
# future_dates = [monthly_sales.index[-1] + DateOffset(months=x) for x in range(1, 13)]
# future_df = pd.DataFrame(index=future_dates, columns=['forecast'])
# future_df['forecast'] = results.forecast(steps=12)

# # Combine historical and forecast data
# combined = pd.concat([monthly_sales, future_df['forecast']], axis=1)
# combined.columns = ['Sales', 'Forecast']

# # Plot the results
# combined.plot(figsize=(10, 5), title='Monthly Sales with Forecast')
# plt.show()

# Fit the ARIMA model
model = ARIMA(monthly_sales, order=(1, 1, 1))
results = model.fit()

# Forecast future sales for historical and future periods
historical_forecast = results.predict(start=monthly_sales.index[1], end=monthly_sales.index[-1],)  # One-step-ahead predictions
future_forecast = results.forecast(steps=12)  # Forecast future values

# Combine historical and forecasted data
forecast_combined = pd.concat([monthly_sales, historical_forecast, future_forecast], axis=1)
forecast_combined.columns = ['Sales', 'Historical Forecast', 'Future Forecast']

# Plot results
forecast_combined[['Sales', 'Historical Forecast', 'Future Forecast']].plot(figsize=(24, 5), title='Monthly Sales with Forecast')
plt.show()
```
